```python
import pandas as pd

# Load the uploaded CSV file
file_path = 'C://Users//acer//OneDrive//Desktop//Bnmit//Projects & Assignments//Cloud Computing Protemics//food.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")

# Calculate the average carbohydrate content per category
avg_carbs = df.groupby('Category')['Data.Carbohydrate'].mean().sort_values()

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=avg_carbs, y=avg_carbs.index, palette="viridis", hue=avg_carbs.index, dodge=False, legend=False)

plt.title('Average Carbohydrate Content per Food Category')
plt.xlabel('Average Carbohydrate Content (g)')
plt.ylabel('Food Category')
plt.show()
```

```python
# Select relevant columns for the heatmap
nutrient_columns = df.columns[df.columns.str.startswith('Data.')]
correlation_matrix = df[nutrient_columns].corr()

# Create a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Heatmap of Nutrients')
plt.show()
```

```python
# Select a subset of nutrient columns for the pair plot
selected_columns = ['Data.Carbohydrate', 'Data.Protein', 'Data.Fat', 'Data.Fiber']
selected_data = df[selected_columns]

# Create a pair plot
sns.pairplot(selected_data, kind='scatter', plot_kws={'alpha':0.6})

plt.suptitle('Pair Plot of Selected Nutrients', y=1.02)
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Data.Vitamins.Vitamin A - RAE', data=df, palette="Set3")

plt.title('Vitamin A Content Across Food Categories')
plt.xlabel('Food Category')
plt.ylabel('Vitamin A - RAE (mcg)')
plt.xticks(rotation=90)
plt.show()
```
